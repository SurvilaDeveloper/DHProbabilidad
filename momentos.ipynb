# ¿Qué es un momento de una distribución?

En estadística y teoría de probabilidades, los **momentos** de una distribución proporcionan una forma de caracterizar la forma y propiedades de una variable aleatoria. Son valores numéricos que describen diversas características como su ubicación, dispersión, asimetría, etc.

---

## 📐 Definición formal

El **n-ésimo momento central** de una distribución se define como:

$$
\mu_n = \mathbb{E}[(X - \mu)^n]
$$

Donde:
- $ X $ es una variable aleatoria.
- $ \mu = \mathbb{E}[X] $ es la media de $ X $.
- $ \mathbb{E}[\cdot] $ es el operador de esperanza o valor esperado.
- $ n $ es el orden del momento.

Cuando no se especifica que es central, se calcula respecto al origen:

$$
\mathbb{E}[X^n]
$$

---

## 🔢 Tipos de momentos

### 1. Primer momento: Media

$$
\mu_1 = \mathbb{E}[X]
$$

Representa el **centro** o valor esperado de la distribución.

---

### 2. Segundo momento: Varianza

$$
\mu_2 = \mathbb{E}[(X - \mu)^2] = \text{Var}(X)
$$

Mide la **dispersión** o variabilidad de la distribución.

---

### 3. Tercer momento: Asimetría (Skewness)

$$
\mu_3 = \mathbb{E}[(X - \mu)^3]
$$

Mide la **asimetría**: si la cola derecha es más larga o más pesada que la izquierda (o viceversa).

---

### 4. Cuarto momento: Curtosis

$$
\mu_4 = \mathbb{E}[(X - \mu)^4]
$$

Mide la **curtosis**: cuán "picuda" o "aplanada" es la distribución comparada con la normal.

---

## 🧮 Propiedades

- **Momento de orden cero**:  
  $$
  \mu_0 = \mathbb{E}[1] = 1
  $$
  Siempre es igual a 1.

- Los momentos describen:
  - La **media** (centro)
  - La **varianza** (anchura)
  - El **sesgo** (asimetría)
  - La **curtosis** (forma de los extremos)

---

## 📊 Ejemplos

### Distribución Normal $ \mathcal{N}(\mu, \sigma^2) $
- $ \mu_1 = \mu $
- $ \mu_2 = \sigma^2 $
- $ \mu_3 = 0 $ (simétrica)
- $ \mu_4 = 3\sigma^4 $

### Distribución Exponencial $ \text{Exp}(\lambda) $
- $ \mu_1 = \frac{1}{\lambda} $
- $ \mu_2 = \frac{2}{\lambda^2} $
- $ \mu_3 = \frac{6}{\lambda^3} $

---

## 📌 Importancia de los momentos

- Se usan en **ajuste de distribuciones** (método de momentos).
- Son claves en **análisis descriptivo**.
- Son usados para establecer **convergencia en distribución** (teorema de los momentos).

---

En resumen, los momentos de una distribución son herramientas fundamentales para analizar el comportamiento de variables aleatorias y comparar distribuciones.
