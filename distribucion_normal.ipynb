# Distribuci√≥n Normal (Gaussiana)

La **distribuci√≥n normal** es una de las distribuciones m√°s importantes en estad√≠stica y probabilidad. Se utiliza para modelar fen√≥menos naturales, errores de medici√≥n, inteligencia, alturas, etc.

---

## üìê Funci√≥n de Densidad de Probabilidad (PDF)

La f√≥rmula de la funci√≥n de densidad de la normal es:

$$
f(x; \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
$$

Donde:
- $ \mu $ es la **media** (centro de la distribuci√≥n).
- $ \sigma $ es la **desviaci√≥n est√°ndar** (ancho o dispersi√≥n).
- $ \sigma^2 $ es la **varianza**.

---

## üß† Propiedades

- Es **sim√©trica** respecto a la media $ \mu $.
- Tiene forma de **campana**.
- La **media, mediana y moda** coinciden.
- El 68% de los valores est√°n dentro de $ \mu \pm \sigma $.
- El 95% est√°n dentro de $ \mu \pm 2\sigma $.
- El 99.7% est√°n dentro de $ \mu \pm 3\sigma $ (regla emp√≠rica o regla 68‚Äì95‚Äì99.7).

---

## üß™ Distribuci√≥n Normal Est√°ndar

Cuando $ \mu = 0 $ y $ \sigma = 1 $, se llama **distribuci√≥n normal est√°ndar** y su PDF es:

$$
f(z) = \frac{1}{\sqrt{2\pi}} e^{ -\frac{z^2}{2} }
$$

Donde $ z = \frac{x - \mu}{\sigma} $ es la variable tipificada.

---

## üìä Usos comunes

- Modelar errores o ruido.
- Teorema Central del L√≠mite: muchas distribuciones tienden a la normal si se promedian suficientes observaciones independientes.
- Inferencia estad√≠stica (intervalos de confianza, pruebas de hip√≥tesis, etc.).

---

## üìå Notaci√≥n

Se denota como:

$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

Por ejemplo, si $ X \sim \mathcal{N}(0, 1) $, entonces es una normal est√°ndar.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.style.use('dark_background')

# Par√°metros de la distribuci√≥n
mu = 0       # media
sigma = 1    # desviaci√≥n est√°ndar
n = 10000    # cantidad de muestras

# Simulaci√≥n de datos
samples = np.random.normal(loc=mu, scale=sigma, size=n)

# Valores para la curva te√≥rica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_theoretical = norm.pdf(x, mu, sigma)

# Graficar histograma y curva te√≥rica
plt.figure(figsize=(10, 5))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue', label='Simulaci√≥n')
plt.plot(x, pdf_theoretical, 'r-', lw=2, label='PDF te√≥rica')

plt.title(f'Distribuci√≥n Normal ‚Äî $\\mu$ = {mu}, $\\sigma$ = {sigma}')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros de la distribuci√≥n
mu = 50       # media
sigma = 10    # desviaci√≥n est√°ndar
n = 10000     # n√∫mero de muestras

# Simulaci√≥n de datos
samples = np.random.normal(loc=mu, scale=sigma, size=n)

# Valores para la curva te√≥rica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_theoretical = norm.pdf(x, mu, sigma)

# Graficar histograma y curva te√≥rica
plt.figure(figsize=(10, 5))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='lightgreen', label='Simulaci√≥n')
plt.plot(x, pdf_theoretical, 'r-', lw=2, label='PDF te√≥rica')

plt.title(f'Distribuci√≥n Normal ‚Äî $\\mu$ = {mu}, $\\sigma$ = {sigma}')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros de la distribuci√≥n no est√°ndar
mu = 50
sigma = 10
n = 10000

# Simulaci√≥n
samples = np.random.normal(loc=mu, scale=sigma, size=n)

# Eje x suficientemente amplio para cubrir ambas curvas
x = np.linspace(-10, mu + 4*sigma, 1000)

# PDFs
pdf_nonstd = norm.pdf(x, mu, sigma)
pdf_std = norm.pdf(x, 0, 1)

# Graficar
plt.figure(figsize=(10, 5))

# Histograma de los datos simulados (normal no est√°ndar)
plt.hist(samples, bins=50, density=True, alpha=0.5, color='skyblue', label='Histograma (simulaci√≥n)')

# Curvas te√≥ricas
plt.plot(x, pdf_nonstd, 'b-', lw=2, label=f'Normal no est√°ndar ($\\mu$={mu}, $\\sigma$={sigma})')
plt.plot(x, pdf_std, 'r--', lw=2, label='Normal est√°ndar ($\\mu$=0, $\\sigma$=1)')

# Est√©tica
plt.title('Histograma vs normales (no est√°ndar y est√°ndar sin escalar)')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, laplace, uniform

# Rango com√∫n para todas las distribuciones
x = np.linspace(-8, 8, 1000)

# PDFs
pdf_norm = norm.pdf(x, loc=0, scale=1)
pdf_laplace = laplace.pdf(x, loc=0, scale=1 / np.sqrt(2))  # ajustamos varianza = 1
pdf_uniform = uniform.pdf(x, loc=-np.sqrt(3), scale=2*np.sqrt(3))  # varianza = 1

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_norm, 'k-', lw=2, label='Normal est√°ndar (mesoc√∫rtica)')
plt.plot(x, pdf_laplace, 'b--', lw=2, label='Laplace (leptoc√∫rtica)')
plt.plot(x, pdf_uniform, 'g-.', lw=2, label='Uniforme (platic√∫rtica)')

# Est√©tica
plt.title('Distribuciones con distinta curtosis (misma varianza)')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
