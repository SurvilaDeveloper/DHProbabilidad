### ✅ Distribución de Poisson

**Problema 2:**  
En una ciudad, se registran en promedio 3 accidentes de tránsito al día.  
¿Cuál es la probabilidad de que **haya más de 4 accidentes** mañana?

Graficar PMF de simulación con 10000 días y X =(0,15) y PMF teórica  
Graficar CDF  
Graficar cuartiles y outliers de la simulación  
**solución:** 0.1847

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetro lambda
lamd = 3  # Tasa de ocurrencia promedio de eventos por intervalo de tiempo

# Simulación
# Generación de datos simulados
size = 10000  # Número de experimentos
datos_simulados = np.random.poisson(lam=lamd, size=size)

#----------------------------------------------------
# Parameters
x = np.arange(0, 15)  # rango de valores para la variable aleatoria x

# función de probabilidad de Poisson
prob_pmf = poisson.pmf(x, mu=3)  # PMF for lambda=3
prob_pmf_escalada= prob_pmf * size  # Escalando la PMF para que coincida con el tamaño de la simulación

# funcion CDF de Poisson
prob_cdf = poisson.cdf(x, mu=3)  # CDF for lambda=3

# probabilidad de 4 accidentes mañana
prob_4_accidents = 1-poisson.cdf(4, mu=3)  # PMF for lambda=3
print (f"Probabilidad de 4 accidentes mañana: {prob_4_accidents:.4f}")


#----------------------------------------------------
# Gráfico de la función de probabilidad de Poisson
plt.figure(figsize=(10, 6))
plt.hist(datos_simulados,bins=x-0.5, alpha=0.5, label='Simulación', color='skyblue', edgecolor='white')
plt.plot(x, prob_pmf_escalada, 'o-', color='red', label='Distribución Teórica', lw=2)
plt.title("Poisson Distribution PMF")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidad")
plt.xticks(np.arange(-1, 16))
plt.xlim(-1, 16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Grafico de la función de distribución acumulativa (CDF) de Poisson
plt.figure(figsize=(10, 6))
plt.step(x,prob_cdf,alpha=0.5,label='probabilidad acumulada', color='blue', lw=2, where='post')
plt.title("Poisson Distribution CDF")
plt.xlabel("Número de eventos")
plt.ylabel("Probabilidad acumulada")
plt.xticks(np.arange(-1, 16))
plt.xlim(-1, 16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()


```
