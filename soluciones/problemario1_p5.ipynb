### ✅ Distribución Exponencial

**Problema 5:**  
El tiempo entre llegadas de clientes a una tienda sigue una distribución exponencial  
con una tasa de $\lambda = 0.25$ clientes por minuto.  
¿Cuál es la probabilidad de que **pasen más de 5 minutos** sin que llegue un cliente?  
Genera 1000 datos aleatorios.
Grafica histograma de los datos, PDF y CDF de la distribución  
Calcula el coeficiente de asimetría y la curtosis.

**✍️ Solución:** P(X > 5) = 0.2865

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
lamb = 0.25
p5 = 1 - expon.cdf(5, scale=1/lamb)
print(f"Problema 5 (Exponencial): P(X > 5) = {p5:.4f}")
```

```python
# Simulación de datos
data = np.random.exponential(scale=1/lamb, size=1000)
x_values = np.linspace(0, np.max(data), 100)

# PDF y CDF de la exponencial
exponencial_pdf = expon.pdf(x_values, scale=1/lamb)
exponencial_cdf = expon.cdf(x_values, scale=1/lamb)

# Asimetría y curtosis
asimetria, curtosis = expon.stats(scale=1/lamb, moments="sk")
print("Asimetría:", asimetria)   # ≈ 2
print("Curtosis:", curtosis)     # ≈ 6

# Gráfico
plt.figure(figsize=(8,5))
plt.title("Distribución Exponencial (λ = 0.25)")
plt.hist(data, bins=50, density=True, alpha=0.3, label="Simulación (1000 datos)")
plt.plot(x_values, exponencial_pdf, label="PDF")
plt.plot(x_values, exponencial_cdf, label="CDF")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
