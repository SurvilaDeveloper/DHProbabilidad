### ✅ Distribución Normal (Gaussiana)

**Problema 3:**  
Las alturas de los adultos en cierta población siguen una distribución normal con  
media $\mu = 1.75$ m y desviación estándar $\sigma = 0.1$ m.  
¿Cuál es la probabilidad de que una persona mida **más de 1.85 m**?  
Graficar PDF y Simulación entre 1.4 y 2.1 metros  
Graficar CDF  
**solución:** 0.1587 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#parametros
mu = 1.75 #media
sigma = 0.1 #desviacion estandar
n = 1000 #numero de muestras

#generar datos
data = np.random.normal(mu, sigma, n)

#Graficar histograma
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histograma de datos')

#Graficar la curva de la normal

x = np.linspace(1.4, 2.1,100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2, label='Curva normal')
plt.title('Histograma y Curva Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid()
plt.show()

# calcular CDF

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
cdf = norm.cdf(x, mu, sigma)
plt.figure(figsize=(10, 6))
plt.plot(x, cdf, 'r-', label='CDF')
plt.title('Función de Distribución Acumulativa (CDF)')
plt.xlabel('Valor')
plt.ylabel('Probabilidad acumulada')
plt.legend()
plt.grid()
plt.show()

# probabilidad de que una persona mida más de 1.85m
prob_mas_185 = 1 - norm.cdf(1.85, mu, sigma)
print(f'Probabilidad de que una persona mida más de 1.85m: {prob_mas_185:.4f}')
```
