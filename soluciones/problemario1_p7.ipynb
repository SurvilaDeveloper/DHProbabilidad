### ✅ Distribución Weibull

**Problema 7:**  
La vida útil (en años) de cierto componente electrónico sigue una distribución Weibull  
con parámetros $k = 1.5$ y $\lambda = 4$.  
¿Cuál es la probabilidad de que el componente **falle antes de los 3 años**?

**✍️ Solución:**  P(X < 3) = 0.4777

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min
```

```python
solucion = weibull_min.cdf(3, c=1.5, scale=4)
solucion2 = 1-weibull_min.sf(3, c=1.5, scale=4)
print(f"Solución (Weibull): P(X < 3) = {solucion:.4f}")
print(f"Solución 2 (Weibull): P(X < 3) = {solucion2:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Parámetros
k = 1.5        # forma
lam = 4        # escala

# Definir el rango de tiempo
t = np.linspace(0, 15, 500)

# Función de densidad de probabilidad (PDF)
pdf = weibull_min.pdf(t, c=k, scale=lam)

# Función de supervivencia (S(t) = 1 - CDF)
survival = weibull_min.sf(t, c=k, scale=lam)

# Crear las gráficas
plt.figure(figsize=(12, 5))

# PDF
plt.subplot(1, 2, 1)
plt.plot(t, pdf, label='PDF', color='blue')
plt.title('Función de Densidad de Probabilidad (Weibull)')
plt.xlabel('Años')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.legend()

# Supervivencia
plt.subplot(1, 2, 2)
plt.plot(t, survival, label='Supervivencia', color='green', linewidth=4)
plt.title('Función de Supervivencia (Weibull)')
plt.xlabel('Años')
plt.ylabel('Probabilidad de sobrevivir')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
