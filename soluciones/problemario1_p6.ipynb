### ✅ Distribución Chi-cuadrada

**Problema 6:**  
Una variable aleatoria $X$ tiene distribución $\chi^2$ con 4 grados de libertad.  
¿Cuál es la probabilidad de que $X$ sea **menor que 2.5**?  

Genera 1000 datos aleatorios.
Grafica histograma de los datos, PDF y CDF de la distribución  
Calcula el coeficiente de asimetría y la curtosis.

**✍️ Solución:** (Chi-cuadrada): P(X < 2.5) = 0.3554

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2 
```

```python
probabilidad = chi2.cdf(x=2.5, df=4)
print(probabilidad)
```

```python
data = np.random.chisquare(df=4,size=1000)

x = np.linspace(np.min(data), np.max(data), 1000)

y_chi2 = chi2.pdf(x=x, df=4)

plt.figure(figsize=(10,5))
plt.hist(data, bins=100, density=True)
plt.plot(x,y_chi2)

plt.show()
```
