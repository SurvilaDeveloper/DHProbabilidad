# Distribución Uniforme Continua

La **distribución uniforme continua** es un tipo de distribución de probabilidad en la que todos los valores dentro de un intervalo específico tienen la misma probabilidad de ocurrir. En otras palabras, cualquier valor dentro del rango dado es igualmente probable.

## Función de Densidad de Probabilidad (PDF)

La **función de densidad de probabilidad (PDF)** de una distribución uniforme continua, definida en el intervalo \([a, b]\), es:

$$
f(x) =
\begin{cases}
\frac{1}{b - a} & \text{si } a \leq x \leq b \\
0 & \text{en otro caso}
\end{cases}
$$

## Propiedades clave

1. **Parámetros**: El intervalo se define por los parámetros \(a\) (el valor mínimo) y \(b\) (el valor máximo).
2. **Media**: La media de la distribución uniforme continua es el punto medio del intervalo, es decir,

   $$
   \mu = \frac{a + b}{2}
   $$

3. **Varianza**: La varianza de la distribución es

   $$
   \sigma^2 = \frac{(b - a)^2}{12}
   $$

4. **Función de Distribución Acumulada (CDF)**: La función de distribución acumulada \(F(x)\) para \(a \leq x \leq b\) es:

   $$
   F(x) = \frac{x - a}{b - a}
   $$

## Ejemplo

Si tienes una variable aleatoria \(X\) que sigue una distribución uniforme continua en el intervalo \([2, 5]\), entonces:

- La densidad de probabilidad será:

  $$
  f(x) = \frac{1}{5 - 2} = \frac{1}{3} \quad \text{para } x \in [2, 5]
  $$

- La media será:

  $$
  \mu = \frac{2 + 5}{2} = 3.5
  $$

- La varianza será:

  $$
  \sigma^2 = \frac{(5 - 2)^2}{12} = \frac{9}{12} = 0.75
  $$

La distribución uniforme continua es bastante simple, pero útil para modelar fenómenos en los que no hay ninguna preferencia por ciertos valores dentro de un intervalo específico.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

# Parámetros de la distribución uniforme
a = 2
b = 5

# Número de muestras para la simulación
n_samples = 10000

# Generar muestras aleatorias de una distribución uniforme continua
samples = np.random.uniform(a, b, n_samples)

# Calcular los valores teóricos de la función de densidad de probabilidad (PDF)
x_values = np.linspace(a, b, 1000)
pdf_values = np.ones_like(x_values) / (b - a)

# Graficar los resultados
plt.figure(figsize=(10, 6))

# Histograma de las muestras simuladas
plt.hist(samples, bins=100, density=True, alpha=0.6, color='g', label='Simulación')

# Línea de la función de densidad teórica
plt.plot(x_values, pdf_values, label='Teórico', color='r', lw=2)

# Agregar títulos y etiquetas
plt.title('Simulación de Distribución Uniforme Continua vs Teórica')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.legend()

# Mostrar el gráfico
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
