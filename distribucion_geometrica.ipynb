# Distribución Geométrica

---

La distribución geométrica modela la **cantidad de ensayos necesarios hasta obtener el primer éxito** en una serie de experimentos independientes, donde cada experimento tiene solo dos posibles resultados: **éxito o fracaso**.

---

## Condiciones principales:
- Cada ensayo tiene la **misma probabilidad de éxito** (denotada como \( p \)).
- Los ensayos son **independientes** entre sí.

---

## Ejemplo sencillo:
Imagina que lanzas una moneda hasta que salga cara:
- Cada lanzamiento tiene probabilidad \( p = 0.5 \) de salir cara.
- ¿Cuántos lanzamientos necesitas para obtener la **primera cara**?
- Eso sigue una distribución geométrica.

---

## Propiedades clave:

- **Variable aleatoria**: Número de ensayos hasta el primer éxito (incluyendo el ensayo exitoso).

- **Media (esperanza)**:

$$
E[X] = \frac{1}{p}
$$

- **Varianza**:

$$
\text{Var}(X) = \frac{1 - p}{p^2}
$$

---

## Fórmula de la distribución geométrica:

La probabilidad de que el primer éxito ocurra en el ensayo número \( k \) es:

$$
P(X = k) = (1 - p)^{k-1} p
$$

donde:
- \( p \) = probabilidad de éxito en un solo ensayo.
- \( k \) = número de ensayos hasta el primer éxito.

**Interpretación**:
- Ocurren \( k-1 \) fracasos seguidos y luego un éxito.

---

## Resumen rápido:

| Concepto | Distribución Geométrica |
|:---|:---|
| **Ensayos** | Hasta el primer éxito |
| **Probabilidad constante** | Sí |
| **Independencia** | Sí |
| **Ejemplo típico** | Lanzar una moneda hasta que salga cara |



```python
# Importando las librerías necesarias
import math
import numpy as np
import matplotlib.pyplot as plt

# Estilo de fondo oscuro
plt.style.use('dark_background')
```

```python
# Parámetros
p = 0.3  # Probabilidad de éxito
size = 10000  # Número de experimentos
```

```python
# P(X=k) = (1-p)^(k-1) * p
# k = número de ensayos hasta el primer éxito
# p = probabilidad de éxito en cada ensayo
def probabilidad_geometrica(k, p):
    return (1 - p) ** (k - 1) * p
```

```python
# Simular la distribución geométrica
datos = np.random.geometric(p, size)
print(datos)

# probabilidad geométrica teorica
vals = [probabilidad_geometrica(i, p) for i in range(1, 16)]
x_vals = np.arange(1, 16)

# Visualizar los resultados
plt.figure(figsize=(12,6))
plt.hist(datos, bins=range(1, 15), color=(0.7,0.7,1.0), density=True, edgecolor='black', align='left', rwidth=0.8, alpha=0.9, label='Simulación')
plt.bar(x_vals,vals, 0.4, color=(1.0,0.3,0.3), edgecolor='black', alpha=0.7, label='Probabilidad Geométrica Teórica')
plt.title(f"Distribución Geométrica (p={p})")
plt.xlabel('Número de ensayos hasta el primer éxito')
plt.ylabel('Frecuencia relativa')
plt.xticks(np.arange(1, 15))
plt.grid(color='gray', linestyle='--', linewidth=0.5, axis='y')
plt.legend()
plt.show()
```
