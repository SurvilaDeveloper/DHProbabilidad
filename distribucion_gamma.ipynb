## Distribución Gamma

La **distribución Gamma** es una distribución de probabilidad continua que se utiliza para modelar tiempos de espera o eventos acumulativos. Es una generalización de la **distribución exponencial** y aparece frecuentemente en teoría de colas, confiabilidad y procesos de Poisson.

### Función de densidad de probabilidad (PDF)

La función de densidad para la distribución Gamma con parámetro de forma $k$ (a veces denotado $\alpha$) y parámetro de escala $\theta$ (a veces denotado $\beta$) es:

$$
f(x; k, \theta) = \frac{1}{\Gamma(k)\theta^k} x^{k-1} e^{-x/\theta}, \quad x > 0
$$

donde:
- $k > 0$ es el **parámetro de forma**,
- $\theta > 0$ es el **parámetro de escala**,
- $\Gamma(k)$ es la función gamma: $\Gamma(k) = \int_0^\infty t^{k-1} e^{-t} dt$.

### Propiedades

- **Media**: $\mathbb{E}[X] = k\theta$
- **Varianza**: $\mathrm{Var}(X) = k\theta^2$
- **Moda**: $(k - 1)\theta$ (para $k \geq 1$)
- **Asimetría**: $\frac{2}{\sqrt{k}}$
- **Curtosis**: $\frac{6}{k}$

### Casos especiales

- Si $k = 1$, la distribución Gamma se convierte en la **distribución exponencial**.
- Si $k$ es un entero positivo, la distribución Gamma se convierte en la **distribución Erlang**.

### Usos comunes

- Modelar el tiempo hasta que ocurren $k$ eventos en un proceso de Poisson.
- En análisis de supervivencia y teoría de colas.
- Como distribución conjugada en inferencia bayesiana (por ejemplo, para la tasa de una exponencial o Poisson).

## Explicación del código de la distribución Gamma

Este código genera una gráfica de la **distribución Gamma** usando los parámetros:

- **Parámetro de forma**: $k = 2.0$
- **Parámetro de escala**: $\theta = 2.0$

### Pasos del código:

1. **Importación de bibliotecas**:
   - `numpy` para manejar arreglos numéricos.
   - `matplotlib.pyplot` para graficar.
   - `scipy.stats.gamma` para acceder a la distribución Gamma.

2. **Definición de parámetros**:
   - Se establecen los valores de $k$ y $\theta$.
   - Se define un rango de valores `x` para evaluar la función de densidad.

3. **Cálculo de la función de densidad de probabilidad (PDF)**:
   - Se usa `gamma.pdf(x, a=k, scale=theta)` para obtener la densidad en cada punto de `x`.

4. **Cálculo de estadísticas**:
   - Se obtiene la **media** y **desviación estándar** con `gamma.mean(...)` y `gamma.std(...)`.

5. **Visualización**:
   - Se traza la curva de la distribución Gamma.
   - Se agregan líneas verticales para marcar:
     - La **media** (línea discontinua).
     - Una **desviación estándar arriba y abajo** de la media (líneas punteadas).

6. **Decoración del gráfico**:
   - Título, etiquetas de ejes, leyenda y cuadrícula.

#### Esta visualización permite entender la forma y características de la distribución Gamma para los parámetros dados. Puedes modificar los valores de $k$ y $\theta$ para explorar otros casos.
---


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

plt.style.use('dark_background')

# Parámetros de la distribución gamma
k = 2.0       # parámetro de forma (shape)
theta = 2.0   # parámetro de escala (scale)

# Crear un rango de valores x
x = np.linspace(0, 20, 1000)

# Calcular la función de densidad de probabilidad
gamma_pdf = gamma.pdf(x, a=k, scale=theta)

# Calcular media y desviación estándar
gamma_mean = gamma.mean(a=k, scale=theta)
gamma_std = gamma.std(a=k, scale=theta)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, gamma_pdf, label=f'Distribución Gamma (k={k}, θ={theta})', color=(0.5,0.5,1.0), linewidth=2)

# Líneas estadísticas
plt.axvline(gamma_mean, color=(0.5,0.5,1.0), linestyle='--', label=f'Media = {gamma_mean:.2f}')
plt.axvline(gamma_mean + gamma_std, color=(0.5,0.5,1.0), linestyle=':', label='±1 Desv. Estándar')
plt.axvline(gamma_mean - gamma_std, color=(0.5,0.5,1.0), linestyle=':')

# Detalles del gráfico
plt.title('Distribución Gamma')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, norm

# Parámetros
k = 2.0        # Forma de la gamma
theta = 2.0    # Escala de la gamma

# Media y desviación estándar de la gamma
gamma_mean = gamma.mean(a=k, scale=theta)
gamma_std = gamma.std(a=k, scale=theta)

# Distribución normal con misma media y desviación estándar
normal_mean = gamma_mean
normal_std = gamma_std

# Rango de valores x
x = np.linspace(-4, gamma_mean + 5*gamma_std, 1000)

# Densidades
gamma_pdf = gamma.pdf(x, a=k, scale=theta)
normal_pdf = norm.pdf(x, loc=normal_mean, scale=normal_std)

# Gráfico
plt.figure(figsize=(10, 6))

plt.plot(x, gamma_pdf, label='Distribución Gamma', color=(0.5,0.5,1.0), linewidth=2)
plt.plot(x, normal_pdf, label='Distribución Normal (misma media y σ)', color='orange', linewidth=2, linestyle='--')

# Líneas de la media
plt.axvline(gamma_mean, color='green', linestyle='--', linewidth=1, label=f'Media = {gamma_mean:.2f}')

# Líneas de ±1 desviación estándar
plt.axvline(gamma_mean + gamma_std, color='green', linestyle=':', linewidth=1, label='±1 Desv. Estándar')
plt.axvline(gamma_mean - gamma_std, color='green', linestyle=':')

# Estética
plt.title('Comparación entre Distribución Gamma y Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.legend(loc='upper right')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
