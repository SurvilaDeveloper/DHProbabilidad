## Distribución Beta

La **distribución Beta** es una distribución de probabilidad continua definida en el intervalo $[0, 1]$. Es muy útil para modelar **proporciones** y **probabilidades**, especialmente cuando se tienen límites conocidos.

---

### 📘 Definición formal

Sea $X \sim \mathrm{Beta}(\alpha, \beta)$ con parámetros $\alpha > 0$ y $\beta > 0$. La función de densidad de probabilidad (PDF) está dada por:

$$
f(x; \alpha, \beta) = \frac{1}{B(\alpha, \beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}, \quad \text{para } x \in [0, 1]
$$

donde $B(\alpha, \beta)$ es la **función beta**, que se puede expresar en términos de la función gamma:

$$
B(\alpha, \beta) = \frac{\Gamma(\alpha) \, \Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

---

### 📊 Propiedades

- **Soporte:** $x \in [0, 1]$
- **Media:**
  $$
  \mathbb{E}[X] = \frac{\alpha}{\alpha + \beta}
  $$
- **Varianza:**
  $$
  \text{Var}(X) = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}
  $$

---

### 🔍 Aplicaciones

- Modelado de proporciones (por ejemplo, el porcentaje de éxito).
- Distribución a priori en estadística bayesiana (como conjugado de la binomial).
- Modelado de incertidumbre cuando la variable está acotada entre 0 y 1.

---

### 🔧 Forma de la distribución según los parámetros:

- Si $\alpha = \beta = 1$: distribución uniforme.
- Si $\alpha > 1$ y $\beta > 1$: forma unimodal (campana).
- Si $\alpha < 1$ y $\beta < 1$: forma en "U".
- Si $\alpha < 1$ y $\beta > 1$: sesgada hacia 0.
- Si $\alpha > 1$ y $\beta < 1$: sesgada hacia 1.

## Código para Graficar Distribuciones Beta

Este código genera y compara **varias distribuciones Beta** variando los parámetros $\alpha$ y $\beta$. Las distribuciones se grafican en el intervalo $[0, 1]$ y se incluyen sus estadísticas principales: **media** y **desviación estándar**.

---

### 🔢 Parámetros usados

Se definen cuatro pares de valores $(\alpha, \beta)$ que representan distintas formas de la distribución Beta:

- $(\alpha = 0.5, \beta = 0.5)$: forma en "U" (alta densidad en los extremos).
- $(\alpha = 2, \beta = 2)$: forma unimodal (simétrica, similar a una campana).
- $(\alpha = 5, \beta = 1)$: sesgada hacia la derecha (concentrada cerca de 1).
- $(\alpha = 1, \beta = 3)$: sesgada hacia la izquierda (concentrada cerca de 0).

---

### ⚙️ Descripción del código

1. **Librerías utilizadas**:
   - `numpy`: para crear una malla de valores de $x$.
   - `matplotlib`: para graficar.
   - `scipy.stats.beta`: para calcular la función de densidad (PDF), media y varianza de la distribución Beta.

2. **Generación del eje x**:
   - Se crea un vector `x` con 1000 puntos equiespaciados en el intervalo $[0, 1]$, donde se evaluará la PDF de la distribución Beta.

3. **Bucle de graficación**:
   - Para cada par $(\alpha, \beta)$:
     - Se crea una instancia de la distribución `beta(a, b)`.
     - Se calcula la **PDF**, la **media**, la **varianza** y la **desviación estándar**.
     - Se grafica la PDF y se muestra la estadística en la leyenda.

4. **Visualización**:
   - El gráfico incluye una leyenda con los valores de $\alpha$, $\beta$, la media ($\mu$) y la desviación estándar ($\sigma$).
   - Se agregan etiquetas y una cuadrícula para facilitar la lectura.

---

### 📌 Objetivo

Este gráfico permite visualizar cómo la distribución Beta cambia según los valores de sus parámetros, y cómo se comportan la media y la desviación estándar en cada caso.

---




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

plt.style.use('dark_background')

# Configuración de parámetros alpha y beta
params = [
    (1,1),    # Uniforme
    (0.5, 0.5),  # Forma en U
    (2, 2),      # Forma unimodal (campana)
    (5, 1),      # Sesgada hacia 1
    (1, 3),      # Sesgada hacia 0
]

x = np.linspace(0, 1, 1000)

# Graficar
plt.figure(figsize=(9, 6))

for a, b in params:
    dist = beta(a, b)
    y = dist.pdf(x)
    mean = dist.mean()
    var = dist.var()
    std = np.sqrt(var)

    label = f"α={a}, β={b}\nμ={mean:.2f}, σ={std:.2f}"
    plt.plot(x, y, label=label)

plt.title("Distribuciones Beta para diferentes α y β")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
