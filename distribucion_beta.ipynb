## Distribuci贸n Beta

La **distribuci贸n Beta** es una distribuci贸n de probabilidad continua definida en el intervalo $[0, 1]$. Es muy 煤til para modelar **proporciones** y **probabilidades**, especialmente cuando se tienen l铆mites conocidos.

###  Definici贸n formal

Sea $X \sim \mathrm{Beta}(\alpha, \beta)$ con par谩metros $\alpha > 0$ y $\beta > 0$. La funci贸n de densidad de probabilidad (PDF) est谩 dada por:

$$
f(x; \alpha, \beta) = \frac{1}{B(\alpha, \beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}, \quad \text{para } x \in [0, 1]
$$

donde $B(\alpha, \beta)$ es la **funci贸n beta**, que se puede expresar en t茅rminos de la funci贸n gamma:

$$
B(\alpha, \beta) = \frac{\Gamma(\alpha) \, \Gamma(\beta)}{\Gamma(\alpha + \beta)}

---

###  Propiedades

- **Soporte:** $x \in [0, 1]$
- **Media:**
  $$
  \mathbb{E}[X] = \frac{\alpha}{\alpha + \beta}
  $$
- **Varianza:**
  $$
  \text{Var}(X) = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}
  $$

---

###  Aplicaciones

- Modelado de proporciones (por ejemplo, el porcentaje de 茅xito).
- Distribuci贸n a priori en estad铆stica bayesiana (como conjugado de la binomial).
- Modelado de incertidumbre cuando la variable est谩 acotada entre 0 y 1.

---

###  Forma de la distribuci贸n seg煤n los par谩metros:

- Si $\alpha = \beta = 1$: distribuci贸n uniforme.
- Si $\alpha > 1$ y $\beta > 1$: forma unimodal (campana).
- Si $\alpha < 1$ y $\beta < 1$: forma en "U".
- Si $\alpha < 1$ y $\beta > 1$: sesgada hacia 0.
- Si $\alpha > 1$ y $\beta < 1$: sesgada hacia 1.

## C贸digo para Graficar Distribuciones Beta

Este c贸digo genera y compara **varias distribuciones Beta** variando los par谩metros $\alpha$ y $\beta$. Las distribuciones se grafican en el intervalo $[0, 1]$ y se incluyen sus estad铆sticas principales: **media** y **desviaci贸n est谩ndar**.

---

###  Par谩metros usados

Se definen cuatro pares de valores $(\alpha, \beta)$ que representan distintas formas de la distribuci贸n Beta:

- $(\alpha = 0.5, \beta = 0.5)$: forma en "U" (alta densidad en los extremos).
- $(\alpha = 2, \beta = 2)$: forma unimodal (sim茅trica, similar a una campana).
- $(\alpha = 5, \beta = 1)$: sesgada hacia la derecha (concentrada cerca de 1).
- $(\alpha = 1, \beta = 3)$: sesgada hacia la izquierda (concentrada cerca de 0).

---

### 锔 Descripci贸n del c贸digo

1. **Librer铆as utilizadas**:
   - `numpy`: para crear una malla de valores de $x$.
   - `matplotlib`: para graficar.
   - `scipy.stats.beta`: para calcular la funci贸n de densidad (PDF), media y varianza de la distribuci贸n Beta.

2. **Generaci贸n del eje x**:
   - Se crea un vector `x` con 1000 puntos equiespaciados en el intervalo $[0, 1]$, donde se evaluar谩 la PDF de la distribuci贸n Beta.

3. **Bucle de graficaci贸n**:
   - Para cada par $(\alpha, \beta)$:
     - Se crea una instancia de la distribuci贸n `beta(a, b)`.
     - Se calcula la **PDF**, la **media**, la **varianza** y la **desviaci贸n est谩ndar**.
     - Se grafica la PDF y se muestra la estad铆stica en la leyenda.

4. **Visualizaci贸n**:
   - El gr谩fico incluye una leyenda con los valores de $\alpha$, $\beta$, la media ($\mu$) y la desviaci贸n est谩ndar ($\sigma$).
   - Se agregan etiquetas y una cuadr铆cula para facilitar la lectura.

---

###  Objetivo

Este gr谩fico permite visualizar c贸mo la distribuci贸n Beta cambia seg煤n los valores de sus par谩metros, y c贸mo se comportan la media y la desviaci贸n est谩ndar en cada caso.

---




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

plt.style.use('dark_background')

# Configuraci贸n de par谩metros alpha y beta
params = [
    (1,1),    # Uniforme
    (0.5, 0.5),  # Forma en U
    (2, 2),      # Forma unimodal (campana)
    (5, 1),      # Sesgada hacia 1
    (1, 3),      # Sesgada hacia 0
]

x = np.linspace(0, 1, 1000)

# Graficar
plt.figure(figsize=(9, 6))

for a, b in params:
    dist = beta(a, b)
    y = dist.pdf(x)
    mean = dist.mean()
    var = dist.var()
    std = np.sqrt(var)

    label = f"伪={a}, 尾={b}\n渭={mean:.2f}, ={std:.2f}"
    plt.plot(x, y, label=label)

plt.title("Distribuciones Beta para diferentes 伪 y 尾")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
