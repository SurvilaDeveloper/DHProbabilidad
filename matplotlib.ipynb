# Tipos de Gr치ficos en Matplotlib

A continuaci칩n se presenta una tabla con los tipos m치s comunes de gr치ficos que se pueden realizar con la librer칤a `matplotlib` en Python, junto con una breve descripci칩n:

| Tipo de gr치fico                     | Funci칩n principal         | Descripci칩n                                                                 |
|------------------------------------|---------------------------|-----------------------------------------------------------------------------|
| L칤nea                              | `plot()`                  | Muestra series de datos o funciones matem치ticas como l칤neas continuas.     |
| Barras verticales / horizontales   | `bar()` / `barh()`        | Representa frecuencias o cantidades mediante barras.                        |
| Histograma                         | `hist()`                  | Distribuci칩n de frecuencias de un conjunto de datos.                        |
| Dispersi칩n                         | `scatter()`               | Muestra puntos individuales en el plano XY.                                |
| Diagrama de caja (boxplot)         | `boxplot()`               | Resume la distribuci칩n con mediana, cuartiles y valores at칤picos.          |
| Gr치fico de viol칤n                  | `violinplot()`            | Variante del boxplot que incluye densidad de probabilidad.                 |
| Gr치fico escalonado                 | `step()`                  | Representa funciones escalonadas (칰til para CDFs).                          |
| Barras de error                    | `errorbar()`              | Muestra valores con sus respectivas barras de incertidumbre.               |
| Gr치fico de tallo y hoja            | `stem()`                  | Representa puntos con l칤neas verticales (칰til para series discretas).      |
| Gr치fico circular (pie chart)       | `pie()`                   | Distribuci칩n de categor칤as como proporciones de un c칤rculo.                |
| Gr치fico de densidad hexagonal      | `hexbin()`                | Densidad de puntos usando celdas hexagonales.                              |
| Campo de vectores                  | `quiver()`                | Representa vectores con flechas (direcci칩n y magnitud).                    |
| L칤neas de flujo                    | `streamplot()`            | Muestra flujos en campos vectoriales.                                      |
| Coordenadas polares                | `polar()`                 | Gr치fico basado en 치ngulo y radio.                                          |
| Gr치fico de radar                   | `plot()` en ejes polares  | Usado para comparar m칰ltiples variables (se necesita configuraci칩n extra). |
| Imagen / matriz de calor           | `imshow()` / `matshow()`  | Representa matrices o im치genes como mapas de color.                        |
| Contornos                          | `contour()` / `contourf()`| Muestra niveles o curvas de igual valor (funciones de dos variables).      |
| Gr치ficos en 3D                     | `Axes3D` de `mpl_toolkits`| Superficies, l칤neas y dispersi칩n en 3D.                                    |

> 游눠 Muchos de estos gr치ficos pueden combinarse en una misma figura o subfiguras (`subplots`), lo que permite construir visualizaciones ricas y personalizadas.



```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.linspace(0, 10, 100)           # 100 puntos entre 0 y 10
y = np.sin(x)                         # funci칩n seno

# Gr치fico
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)

# Decoraci칩n del gr치fico
plt.title('Gr치fico de L칤nea con plot()')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [23, 17, 35, 29]

# Crear gr치fico de barras
plt.bar(categorias, valores, color='skyblue', edgecolor='black')

# Decoraci칩n del gr치fico
plt.title('Gr치fico de Barras con bar()')
plt.xlabel('Categor칤as')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios: 1000 muestras de una distribuci칩n normal
data = np.random.normal(loc=0, scale=1, size=1000)

# Crear histograma
plt.hist(data, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)

# Decoraci칩n del gr치fico
plt.title('Histograma con hist()')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + 1 + np.random.normal(0, 0.1, size=100)  # y 곋 2x + 1 con algo de ruido

# Crear gr치fico de dispersi칩n
plt.scatter(x, y, color='teal', alpha=0.7, edgecolor='black')

# Decoraci칩n del gr치fico
plt.title('Gr치fico de Dispersi칩n con scatter()')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(1)
grupo1 = np.random.normal(loc=50, scale=10, size=100)
grupo2 = np.random.normal(loc=60, scale=15, size=100)
grupo3 = np.random.normal(loc=55, scale=5, size=100)

# Agrupar los datos
datos = [grupo1, grupo2, grupo3]

# Crear boxplot
plt.boxplot(datos, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'], patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='black'))

# Decoraci칩n del gr치fico
plt.title('Boxplot con boxplot()')
plt.ylabel('Valor')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(2)
grupo1 = np.random.normal(loc=40, scale=5, size=100)
grupo2 = np.random.normal(loc=50, scale=10, size=100)
grupo3 = np.random.normal(loc=60, scale=7, size=100)

# Agrupar datos
datos = [grupo1, grupo2, grupo3]

# Crear gr치fico de viol칤n
plt.violinplot(dataset=datos, showmeans=True, showmedians=True)

# Personalizaci칩n del gr치fico
plt.title('Gr치fico de Viol칤n con violinplot()')
plt.xticks([1, 2, 3], ['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.ylabel('Valor')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos: distribuci칩n discreta de ejemplo
x = np.arange(0, 6)
p = [0.1, 0.2, 0.3, 0.2, 0.15, 0.05]  # Probabilidades
cdf = np.cumsum(p)  # CDF acumulada

# Crear gr치fico escalonado
plt.step(x, cdf, where='post', color='purple', linewidth=2)

# Decoraci칩n del gr치fico
plt.title('Gr치fico Escalonado con step()')
plt.xlabel('Valor')
plt.ylabel('Probabilidad acumulada (CDF)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.ylim(0, 1.05)
plt.xticks(x)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.linspace(0, 10, 10)
y = np.sin(x)
error = 0.1 + 0.2 * np.sqrt(x)  # Error creciente

# Crear gr치fico con barras de error
plt.errorbar(x, y, yerr=error, fmt='o', color='teal', ecolor='orange', elinewidth=2, capsize=5)

# Decoraci칩n del gr치fico
plt.title('Gr치fico con barras de error usando errorbar()')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
X, Y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))

# Definir las componentes del campo vectorial (direcci칩n y magnitud)
U = np.cos(X)  # Componente horizontal
V = np.sin(Y)  # Componente vertical

# Crear gr치fico de vectores
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color='r')

# Decoraci칩n del gr치fico
plt.title('Gr치fico de Vectores con quiver()')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim([0, 10])
plt.ylim([0, 10])

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
Y, X = np.mgrid[0:10:100j, 0:10:100j]

# Definir las componentes del campo vectorial (direcci칩n y magnitud)
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Crear gr치fico de l칤neas de corriente
plt.streamplot(X, Y, U, V, color=np.sqrt(U**2 + V**2), linewidth=2, cmap='autumn')

# Decoraci칩n del gr치fico
plt.title('Gr치fico de L칤neas de Corriente con streamplot()')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 10)
plt.ylim(0, 10)

plt.colorbar(label='Magnitud del vector')

plt.show()
```

##### La principal diferencia entre imshow() y matshow() es que matshow() est치 espec칤ficamente dise침ado para mostrar matrices 2D, y ajusta autom치ticamente las dimensiones de la matriz para ocupar todo el espacio disponible.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Definir la funci칩n z = f(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gr치fico de contornos
plt.contour(X, Y, Z, 20, cmap='viridis')  # 20 niveles de contorno

# A침adir barra de color
plt.colorbar()

# Decoraci칩n del gr치fico
plt.title('Gr치fico de Contornos con contour()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Definir la funci칩n z = f(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gr치fico de contornos llenos
plt.contourf(X, Y, Z, 20, cmap='viridis')  # 20 niveles de contorno llenos

# A침adir barra de color
plt.colorbar()

# Decoraci칩n del gr치fico
plt.title('Gr치fico de Contornos Llenos con contourf()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crear una malla de puntos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Definir la funci칩n z = f(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear una figura para 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar superficie 3D
ax.plot_surface(X, Y, Z, cmap='viridis')

# Decoraci칩n del gr치fico
ax.set_title('Gr치fico 3D con Axes3D')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Mostrar el gr치fico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura con 2 subplots (1 fila, 2 columnas)
plt.figure(figsize=(10, 4))

# Primer subplot
plt.subplot(1, 2, 1)  # (nfilas, ncolumnas, 칤ndice)
plt.plot(x, y1, color='blue')
plt.title("Seno")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Segundo subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='green')
plt.title("Coseno")
plt.xlabel("x")
plt.ylabel("cos(x)")

# Ajustar el layout para que no se solapen los elementos
plt.tight_layout()

# Mostrar la figura
plt.show()
```
