# Tipos de Gráficos en Matplotlib

A continuación se presenta una tabla con los tipos más comunes de gráficos que se pueden realizar con la librería `matplotlib` en Python, junto con una breve descripción:

| Tipo de gráfico                     | Función principal         | Descripción                                                                 |
|------------------------------------|---------------------------|-----------------------------------------------------------------------------|
| Línea                              | `plot()`                  | Muestra series de datos o funciones matemáticas como líneas continuas.     |
| Barras verticales / horizontales   | `bar()` / `barh()`        | Representa frecuencias o cantidades mediante barras.                        |
| Histograma                         | `hist()`                  | Distribución de frecuencias de un conjunto de datos.                        |
| Dispersión                         | `scatter()`               | Muestra puntos individuales en el plano XY.                                |
| Diagrama de caja (boxplot)         | `boxplot()`               | Resume la distribución con mediana, cuartiles y valores atípicos.          |
| Gráfico de violín                  | `violinplot()`            | Variante del boxplot que incluye densidad de probabilidad.                 |
| Gráfico escalonado                 | `step()`                  | Representa funciones escalonadas (útil para CDFs).                          |
| Barras de error                    | `errorbar()`              | Muestra valores con sus respectivas barras de incertidumbre.               |
| Gráfico de tallo y hoja            | `stem()`                  | Representa puntos con líneas verticales (útil para series discretas).      |
| Gráfico circular (pie chart)       | `pie()`                   | Distribución de categorías como proporciones de un círculo.                |
| Gráfico de densidad hexagonal      | `hexbin()`                | Densidad de puntos usando celdas hexagonales.                              |
| Campo de vectores                  | `quiver()`                | Representa vectores con flechas (dirección y magnitud).                    |
| Líneas de flujo                    | `streamplot()`            | Muestra flujos en campos vectoriales.                                      |
| Coordenadas polares                | `polar()`                 | Gráfico basado en ángulo y radio.                                          |
| Gráfico de radar                   | `plot()` en ejes polares  | Usado para comparar múltiples variables (se necesita configuración extra). |
| Imagen / matriz de calor           | `imshow()` / `matshow()`  | Representa matrices o imágenes como mapas de color.                        |
| Contornos                          | `contour()` / `contourf()`| Muestra niveles o curvas de igual valor (funciones de dos variables).      |
| Gráficos en 3D                     | `Axes3D` de `mpl_toolkits`| Superficies, líneas y dispersión en 3D.                                    |

> 💡 Muchos de estos gráficos pueden combinarse en una misma figura o subfiguras (`subplots`), lo que permite construir visualizaciones ricas y personalizadas.



```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.linspace(0, 10, 100)           # 100 puntos entre 0 y 10
y = np.sin(x)                         # función seno

# Gráfico
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)

# Decoración del gráfico
plt.title('Gráfico de Línea con plot()')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [23, 17, 35, 29]

# Crear gráfico de barras
plt.bar(categorias, valores, color='skyblue', edgecolor='black')

# Decoración del gráfico
plt.title('Gráfico de Barras con bar()')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['Python', 'JavaScript', 'C++', 'Java']
valores = [45, 30, 15, 25]

# Crear gráfico de barras horizontales
plt.barh(categorias, valores, color='lightgreen', edgecolor='black')

# Decoración del gráfico
plt.title('Gráfico de Barras Horizontales con barh()')
plt.xlabel('Popularidad (%)')
plt.ylabel('Lenguajes de Programación')
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios: 1000 muestras de una distribución normal
data = np.random.normal(loc=0, scale=1, size=1000)

# Crear histograma
plt.hist(data, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)

# Decoración del gráfico
plt.title('Histograma con hist()')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + 1 + np.random.normal(0, 0.1, size=100)  # y ≈ 2x + 1 con algo de ruido

# Crear gráfico de dispersión
plt.scatter(x, y, color='teal', alpha=0.7, edgecolor='black')

# Decoración del gráfico
plt.title('Gráfico de Dispersión con scatter()')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(1)
grupo1 = np.random.normal(loc=50, scale=10, size=100)
grupo2 = np.random.normal(loc=60, scale=15, size=100)
grupo3 = np.random.normal(loc=55, scale=5, size=100)

# Agrupar los datos
datos = [grupo1, grupo2, grupo3]

# Crear boxplot
plt.boxplot(datos, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'], patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='black'))

# Decoración del gráfico
plt.title('Boxplot con boxplot()')
plt.ylabel('Valor')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(2)
grupo1 = np.random.normal(loc=40, scale=5, size=100)
grupo2 = np.random.normal(loc=50, scale=10, size=100)
grupo3 = np.random.normal(loc=60, scale=7, size=100)

# Agrupar datos
datos = [grupo1, grupo2, grupo3]

# Crear gráfico de violín
plt.violinplot(dataset=datos, showmeans=True, showmedians=True)

# Personalización del gráfico
plt.title('Gráfico de Violín con violinplot()')
plt.xticks([1, 2, 3], ['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.ylabel('Valor')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos: distribución discreta de ejemplo
x = np.arange(0, 6)
p = [0.1, 0.2, 0.3, 0.2, 0.15, 0.05]  # Probabilidades
cdf = np.cumsum(p)  # CDF acumulada

# Crear gráfico escalonado
plt.step(x, cdf, where='post', color='purple', linewidth=2)

# Decoración del gráfico
plt.title('Gráfico Escalonado con step()')
plt.xlabel('Valor')
plt.ylabel('Probabilidad acumulada (CDF)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.ylim(0, 1.05)
plt.xticks(x)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.linspace(0, 10, 10)
y = np.sin(x)
error = 0.1 + 0.2 * np.sqrt(x)  # Error creciente

# Crear gráfico con barras de error
plt.errorbar(x, y, yerr=error, fmt='o', color='teal', ecolor='orange', elinewidth=2, capsize=5)

# Decoración del gráfico
plt.title('Gráfico con barras de error usando errorbar()')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.linspace(0, 2 * np.pi, 10)  # 10 puntos entre 0 y 2pi
y = np.sin(x)

# Crear gráfico de tallo y hoja
plt.stem(x, y, basefmt=" ")

# Decoración del gráfico
plt.title('Gráfico de Tallo y Hoja con stem()')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()

```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
labels = ['Python', 'JavaScript', 'C++', 'Java', 'Ruby']
sizes = [40, 30, 15, 10, 5]  # Porcentaje de cada lenguaje
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen', 'lightpink']
explode = (0.1, 0, 0, 0, 0)  # Resaltar la primera sección

# Crear gráfico de pastel
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Decoración del gráfico
plt.title('Distribución de Lenguajes de Programación')

plt.axis('equal')  # Asegura que el gráfico de pastel sea circular

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
X, Y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))

# Definir las componentes del campo vectorial (dirección y magnitud)
U = np.cos(X)  # Componente horizontal
V = np.sin(Y)  # Componente vertical

# Crear gráfico de vectores
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color='r')

# Decoración del gráfico
plt.title('Gráfico de Vectores con quiver()')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim([0, 10])
plt.ylim([0, 10])

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
Y, X = np.mgrid[0:10:100j, 0:10:100j]

# Definir las componentes del campo vectorial (dirección y magnitud)
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Crear gráfico de líneas de corriente
plt.streamplot(X, Y, U, V, color=np.sqrt(U**2 + V**2), linewidth=2, cmap='autumn')

# Decoración del gráfico
plt.title('Gráfico de Líneas de Corriente con streamplot()')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 10)
plt.ylim(0, 10)

plt.colorbar(label='Magnitud del vector')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
theta = np.linspace(0, 2 * np.pi, 100)  # Ángulos de 0 a 2pi
r = np.abs(np.sin(theta) * np.cos(theta))  # Radio en función de theta

# Crear gráfico polar
plt.polar(theta, r, color='b', linestyle='-', linewidth=2)

# Decoración del gráfico
plt.title('Gráfico Polar con polar()')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
theta = np.linspace(0, 2 * np.pi, 100)  # Ángulos de 0 a 2pi
r = np.abs(np.sin(theta) * np.cos(theta))  # Radio en función de theta

# Crear figura y ejes
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# Graficar en ejes polares
ax.plot(theta, r, color='r', linewidth=2)

# Decoración del gráfico
ax.set_title('Gráfico Polar usando plot()')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una matriz de datos 2D (imagen aleatoria)
data = np.random.rand(10, 10)

# Crear gráfico de la matriz
plt.matshow(data, cmap='plasma')

# Añadir barra de color
plt.colorbar()

# Decoración del gráfico
plt.title('Gráfico con matshow()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```

##### La principal diferencia entre imshow() y matshow() es que matshow() está específicamente diseñado para mostrar matrices 2D, y ajusta automáticamente las dimensiones de la matriz para ocupar todo el espacio disponible.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Definir la función z = f(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gráfico de contornos
plt.contour(X, Y, Z, 20, cmap='viridis')  # 20 niveles de contorno

# Añadir barra de color
plt.colorbar()

# Decoración del gráfico
plt.title('Gráfico de Contornos con contour()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Definir la función z = f(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gráfico de contornos llenos
plt.contourf(X, Y, Z, 20, cmap='viridis')  # 20 niveles de contorno llenos

# Añadir barra de color
plt.colorbar()

# Decoración del gráfico
plt.title('Gráfico de Contornos Llenos con contourf()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crear una malla de puntos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Definir la función z = f(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear una figura para 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar superficie 3D
ax.plot_surface(X, Y, Z, cmap='viridis')

# Decoración del gráfico
ax.set_title('Gráfico 3D con Axes3D')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura con 2 subplots (1 fila, 2 columnas)
plt.figure(figsize=(10, 4))

# Primer subplot
plt.subplot(1, 2, 1)  # (nfilas, ncolumnas, índice)
plt.plot(x, y1, color='blue')
plt.title("Seno")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Segundo subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='green')
plt.title("Coseno")
plt.xlabel("x")
plt.ylabel("cos(x)")

# Ajustar el layout para que no se solapen los elementos
plt.tight_layout()

# Mostrar la figura
plt.show()
```
