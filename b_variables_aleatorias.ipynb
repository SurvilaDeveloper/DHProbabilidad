# Variables Aleatorias

## 1. Variables Aleatorias

Una **variable aleatoria** es una transformación:

$$
X: \Omega \longrightarrow \mathbb{R}
$$

donde $\Omega$ es el espacio muestral y $\mathbb{R}$ el conjunto de los números reales.

El valor que toma una variable aleatoria está determinado por el resultado de un experimento aleatorio.

---

## 2. Función de Distribución

La **función de distribución** de una variable aleatoria \( X \) es una función:

$$
F_X(x) = P(X \leq x)
$$

que asigna a cada número real \( x \) la probabilidad de que \( X \) sea menor o igual que \( x \).

**Notas:**
- La distribución de probabilidad queda completamente especificada por $( F_X(x) )$.
- El valor de $( F_X(x) )$ representa la acumulación de probabilidades hasta \( x \).

---

## 3. Variables Aleatorias Discretas

Una variable aleatoria \( X \) es **discreta** si toma un conjunto finito o numerable de valores posibles.

**Ejemplos:**
- Lanzamiento de un dado:

$$
X \in \{1, 2, 3, 4, 5, 6\}
$$

- Número de hijos en una familia:

$$
X \in \{0, 1, 2, 3, \dots\}
$$

- Número de llamadas en una hora:

$$
X \in \{0, 1, 2, 3, \dots\}
$$

---

### Distribución Puntual de Probabilidad

La **distribución puntual de probabilidad** de una v.a. discreta \( X \) es:

$$
P(X = x_i)
$$

donde se cumple:

- **No negatividad**:

$$
P(X = x_i) \geq 0 \quad \text{para todo } i
$$

- **Suma total igual a uno**:

$$
\sum_{i} P(X = x_i) = 1
$$


---

## 4. Variables Aleatorias Continuas

Una variable aleatoria \( X \) es **continua** si su función de distribución acumulativa \( F_X(x) \) es continua en \( x \).

---

### Función de Densidad de Probabilidad

La **función de densidad** \( f(x) \) cumple:

$$
F(x) = \int_{-\infty}^{x} f(t) \, dt
$$

donde:

- $( f(x) )$ describe la **probabilidad relativa** de los resultados.
- La **función de distribución** $( F(x) )$ es la integral acumulada de $( f(x) )$.

---






# Función de Distribución

La **función de distribución acumulada** (FDA) de una variable aleatoria \( X \) es una función que asigna a cada número real \( x \) la probabilidad de que \( X \) tome un valor menor o igual que \( x \).

Se denota como:

$$
F_X(x) = P(X \leq x)
$$

---

## Propiedades de la Función de Distribución

1. **Valores entre 0 y 1**:

$$
0 \leq F_X(x) \leq 1 \quad \forall x \in \mathbb{R}
$$

2. **Función monótonamente no decreciente**:

Si \( a < b \), entonces:

$$
F_X(a) \leq F_X(b)
$$

3. **Límites**:

$$
\lim_{x \to -\infty} F_X(x) = 0
$$

$$
\lim_{x \to +\infty} F_X(x) = 1
$$

4. **Continuidad por la derecha**:

$$
\lim_{h \to 0^+} F_X(x+h) = F_X(x)
$$

---

## Interpretación

- Para cualquier número real \( x \), \( F_X(x) \) representa la **probabilidad acumulada** de que la variable aleatoria \( X \) tome un valor **menor o igual** que \( x \).
- La gráfica de \( F_X(x) \) es una curva o una función escalonada (en el caso discreto) que siempre sube o se mantiene constante.

---

## Tipos de Variables y sus Funciones de Distribución

### Variables Aleatorias Discretas

- La función \( F_X(x) \) es una **función escalonada**.
- Ejemplo: número de caras al lanzar dos monedas.

### Variables Aleatorias Continuas

- La función \( F_X(x) \) es **suave y continua**.
- Ejemplo: tiempo que tarda un auto en llegar a un destino.

---

## Relación con la Función de Densidad

Si \( X \) es una variable aleatoria continua con función de densidad \( f_X(x) \), entonces:

$$
F_X(x) = \int_{-\infty}^{x} f_X(t) \, dt
$$

y además:

$$
f_X(x) = \frac{d}{dx} F_X(x)
$$

---

# Ejemplo: Función de Distribución Discreta

Supongamos el lanzamiento de un dado equilibrado.  
La variable \( X \) es el número que sale al lanzar el dado. Su función de distribución \( F_X(x) \) sería:

$$
F_X(x) =
\begin{cases}
0 & \text{si } x < 1 \\
\frac{1}{6} & \text{si } 1 \leq x < 2 \\
\frac{2}{6} & \text{si } 2 \leq x < 3 \\
\frac{3}{6} & \text{si } 3 \leq x < 4 \\
\frac{4}{6} & \text{si } 4 \leq x < 5 \\
\frac{5}{6} & \text{si } 5 \leq x < 6 \\
1 & \text{si } x \geq 6
\end{cases}
$$

```python
import matplotlib.pyplot as plt

def funcion_distribucionFDA(x):
    if x < 1:
        return 0.0
    elif x < 2:
        return 1/6
    elif x < 3:
        return 1/3
    elif x < 4:
        return 1/2
    elif x < 5:
        return 2/3
    elif x < 6:
        return 5/6
    else:
        return 1.0

probabilidad_1 = funcion_distribucionFDA(1)
probabilidad_2 = funcion_distribucionFDA(2) - probabilidad_1
probabilidad_3 = funcion_distribucionFDA(3) - (probabilidad_1 + probabilidad_2)
probabilidad_4 = funcion_distribucionFDA(4) - (probabilidad_1 + probabilidad_2 + probabilidad_3)
probabilidad_5 = funcion_distribucionFDA(5) - (probabilidad_1 + probabilidad_2 + probabilidad_3 + probabilidad_4)
probabilidad_6 = funcion_distribucionFDA(6) - (probabilidad_1 + probabilidad_2 + probabilidad_3 + probabilidad_4 + probabilidad_5)

print(f"Probabilidad de salga el número 1: {probabilidad_1:.2f}")
print(f"Probabilidad de salga el número 2: {probabilidad_2:.2f}")
print(f"Probabilidad de salga el número 3: {probabilidad_3:.2f}")
print(f"Probabilidad de salga el número 4: {probabilidad_4:.2f}")
print(f"Probabilidad de salga el número 5: {probabilidad_5:.2f}")
print(f"Probabilidad de salga el número 6: {probabilidad_6:.2f}")
```

```python
x_values = [0,1, 2, 3, 4, 5, 6, 7]
y_values = [funcion_distribucionFDA(x) for x in x_values]
plt.step(x_values, y_values,where='post', marker='o', linestyle='-', label='FDA(x)')
plt.title('Función de Distribución Acumulada - Lanzamiento de un Dado')
plt.xlabel('x')
plt.ylabel('FDA(x)')
plt.grid()
plt.xticks(x_values)
plt.yticks([0, 1/6, 1/3, 1/2, 2/3, 5/6, 1], ['0', '1/6', '1/3', '1/2', '2/3', '5/6', '1'])
plt.legend()
plt.show()
```

# Distribución Puntual de Probabilidad - Lanzamiento de un Dado

La **distribución puntual de probabilidad** describe las probabilidades asociadas a cada posible valor de una variable aleatoria discreta.

En el lanzamiento de un dado equilibrado:

- Valores posibles: $$ X \in \{1, 2, 3, 4, 5, 6\} $$
- Probabilidad de cada valor:

$$
P(X = x) = \frac{1}{6} \quad \text{para cada } x \in \{1,2,3,4,5,6\}
$$

Representación de la distribución:

| $$ x $$ | 1 | 2 | 3 | 4 | 5 | 6 |
|:------:|:--:|:--:|:--:|:--:|:--:|:--:|
| $$ P(X = x) $$ | $$ \frac{1}{6} $$ | $$ \frac{1}{6} $$ | $$ \frac{1}{6} $$ | $$ \frac{1}{6} $$ | $$ \frac{1}{6} $$ | $$ \frac{1}{6} $$ |

---
A continuación, se muestra el gráfico de la distribución puntual de probabilidad.


```python
import matplotlib.pyplot as plt

# Valores posibles
x = [1, 2, 3, 4, 5, 6]
# Probabilidades correspondientes
p = [1/6] * 6

# Crear gráfico
plt.figure(figsize=(8, 5))
plt.stem(x, p)
plt.title('Distribución Puntual de Probabilidad - Dado')
plt.xlabel('Valor de X')
plt.ylabel('P(X=x)')
plt.xticks(x)
plt.ylim(0, 0.3)
plt.grid(True)
plt.show()
```
