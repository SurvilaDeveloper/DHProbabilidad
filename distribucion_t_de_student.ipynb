## Distribución t de Student

La **distribución t de Student** es una distribución de probabilidad continua que se utiliza principalmente para realizar inferencias estadísticas cuando el tamaño de la muestra es pequeño y la varianza poblacional es desconocida.

Se parece a la distribución normal, pero con **colas más pesadas**, lo que le da mayor robustez frente a valores extremos. A medida que aumentan los grados de libertad, la distribución t se aproxima a la normal estándar.

### Función de densidad de probabilidad (PDF)

La función de densidad de probabilidad para la distribución t con $\nu$ grados de libertad es:

$$
f(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \, \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$

donde:
- $\Gamma$ es la función gamma (una generalización del factorial),
- $\nu$ son los grados de libertad,
- $t$ es la variable aleatoria.

### Propiedades

- **Media**: $0$ (si $\nu > 1$)
- **Varianza**: $\frac{\nu}{\nu - 2}$ (si $\nu > 2$; indefinida si $1 < \nu \leq 2$)
- **Simetría**: La distribución es simétrica respecto a la media.
- **Curtosis**: Más pesada que la normal, especialmente con pocos grados de libertad.

### Usos comunes

- Pruebas t para una muestra o dos muestras (comparar medias).
- Intervalos de confianza para la media cuando no se conoce la desviación estándar poblacional.
- Regresión lineal (para estimar la significancia de los coeficientes).

A medida que $\nu \to \infty$, la distribución t converge hacia la **distribución normal estándar**.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use('dark_background')

# Parámetros
mu = 0
sigma = 1
df = 3  # grados de libertad para la t de Student

# Rango de valores para el eje x
x = np.linspace(-5, 5, 1000)

# Funciones de densidad de probabilidad
normal_pdf = stats.norm.pdf(x, mu, sigma)
t_pdf = stats.t.pdf(x, df)

# Estadísticas normales
normal_mean = mu
normal_std_dev = sigma
normal_variance = sigma**2

# Estadísticas t de Student
t_mean = stats.t.mean(df)
t_variance = stats.t.var(df)
t_std_dev = np.sqrt(t_variance)

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Graficar distribuciones
plt.plot(x, normal_pdf, label="Distribución Normal", color=(0.5,0.5,1.0), linewidth=2)
plt.plot(x, t_pdf, label=f'Distribución t-Student (df={df})', color='orange', linewidth=2)

# Líneas estadísticas para la normal
plt.axvline(normal_mean, color=(0.5,0.5,1.0), linestyle='--', label=f'Media Normal = {normal_mean:.2f}')
plt.axvline(normal_mean + normal_std_dev, color=(0.5,0.5,1.0), linestyle=':', label='±1 Desv. Estándar Normal')
plt.axvline(normal_mean - normal_std_dev, color=(0.5,0.5,1.0), linestyle=':')

# Líneas estadísticas para la t de Student
plt.axvline(t_mean, color='orange', linestyle='--', label=f'Media t-Student = {t_mean:.2f}')
plt.axvline(t_mean + t_std_dev, color='orange', linestyle=':', label='±1 Desv. Estándar t-Student')
plt.axvline(t_mean - t_std_dev, color='orange', linestyle=':')

# Detalles del gráfico
plt.title('Distribución Normal vs t de Student')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend(loc='upper right')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
