# Distribuci√≥n Hipergeom√©trica

---

## ¬øQu√© es?

La distribuci√≥n hipergeom√©trica modela el **n√∫mero de √©xitos** en una muestra de tama√±o fijo, **extra√≠da sin reemplazo** de una poblaci√≥n finita que contiene un n√∫mero determinado de √©xitos y fracasos.

üìö Es como sacar cartas de una baraja sin volver a meterlas: lo que sacaste afecta la siguiente extracci√≥n.

---

## Ejemplo sencillo

Imagin√°:

- Una bolsa con 10 bolas, 4 son rojas y 6 son azules.
- Sac√°s 3 bolas al azar **sin devolverlas**.

**Pregunta**: ¬øCu√°l es la probabilidad de que saques exactamente 2 bolas rojas?

üîé ¬°Eso sigue una distribuci√≥n hipergeom√©trica!

---

## F√≥rmula de la Distribuci√≥n

La probabilidad de observar exactamente \( k \) √©xitos es:

$$
P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}
$$

donde:

- \( N \) = Tama√±o de la poblaci√≥n total.
- \( K \) = N√∫mero total de √©xitos en la poblaci√≥n.
- \( n \) = Tama√±o de la muestra.
- \( k \) = N√∫mero de √©xitos observados en la muestra.
- $ \binom{a}{b} $ = Coeficiente binomial ("a sobre b"). La f√≥rmula del coeficiente binomial es:

$$
\binom{a}{b} = \frac{a!}{b!(a - b)!}
$$



---

## Propiedades Clave

| Propiedad              | F√≥rmula |
|:------------------------|:--------|
| Esperanza (media)        | $$ E[X] = n \frac{K}{N} $$ |
| Varianza                 | $$ \mathrm{Var}(X) = n \frac{K}{N} \left(1 - \frac{K}{N}\right) \frac{N-n}{N-1} $$ |

---

## Resumen r√°pido

| Concepto               | Detalle |
|:------------------------|:--------|
| Tipo de muestra          | Sin reemplazo |
| Tama√±o de poblaci√≥n      | Finito |
| Variable aleatoria       | N√∫mero de √©xitos |
| Independencia de ensayos | No (porque no se reemplaza) |
| Ejemplo t√≠pico           | Sacar cartas, bolas, personas |

# Distribuci√≥n Hipergeom√©trica - Simulaci√≥n y Gr√°fica

---

## Librer√≠as
Importamos:
- **numpy**: para manejar arrays y simulaciones.
- **matplotlib**: para graficar.
- **scipy.stats.hypergeom**: para trabajar con la distribuci√≥n hipergeom√©trica.

---

## Par√°metros

- \( N \): Tama√±o de la poblaci√≥n total (en el ejemplo, 50 elementos).
- \( K \): N√∫mero de √©xitos en la poblaci√≥n (20 √©xitos).
- \( n \): Tama√±o de la muestra extra√≠da (10 elementos).

---

## Procedimiento

1. Creamos la variable aleatoria hipergeom√©trica usando `hypergeom(N, K, n)`.

2. Definimos el rango de valores posibles para el n√∫mero de √©xitos \( x \) (de 0 hasta \( n \)).

3. Calculamos la **funci√≥n de masa de probabilidad** (PMF) usando `.pmf(x)`, que da la probabilidad te√≥rica para cada valor de \( x \).

4. Realizamos 10.000 simulaciones de la distribuci√≥n con `.rvs(size=10000)` para obtener datos emp√≠ricos.

---

## Gr√°fico

- El **histograma** muestra los resultados de las simulaciones (datos observados).
- La **l√≠nea roja** con puntos marca las probabilidades te√≥ricas exactas.

> Se puede ver c√≥mo la simulaci√≥n se aproxima bastante a la distribuci√≥n te√≥rica.

---

## Resumen

Esta simulaci√≥n permite entender c√≥mo la distribuci√≥n hipergeom√©trica modela:
- La probabilidad de obtener cierto n√∫mero de √©xitos.
- **Extracciones sin reemplazo**, donde el resultado de un ensayo afecta al siguiente.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Par√°metros
N = 50    # Tama√±o total de la poblaci√≥n
K = 20    # N√∫mero de √©xitos en la poblaci√≥n
n = 10    # Tama√±o de la muestra

# Crear la variable aleatoria hipergeom√©trica
rv = hypergeom(N, K, n)

# Valores posibles
x = np.arange(0, n+1)
print(x)

# Probabilidades
pmf = rv.pmf(x)
print(pmf)

# Simulaci√≥n
simulaciones = rv.rvs(size=10000)
print(simulaciones)

# Graficamos
plt.style.use('dark_background')  # Fondo oscuro opcional
plt.figure(figsize=(10,6))
plt.hist(simulaciones, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulaci√≥n')
plt.plot(x, pmf, 'o-', color='red', label='Distribuci√≥n Te√≥rica')
plt.title('Distribuci√≥n Hipergeom√©trica')
plt.xlabel('N√∫mero de √©xitos en la muestra')
plt.ylabel('Probabilidad')
plt.xticks(x)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
